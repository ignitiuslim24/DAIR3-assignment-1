```python
import pandas as pd

df = pd.read_excel("mtcars.xlsx", engine='openpyxl')
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset (if not already loaded)
# df = pd.read_excel("your_file.xlsx", engine=''openpyxl'')

# Basic check for column names
print(df.columns)

# Replace ''car_type'' with your actual column name for car type if different
# Replace ''mpg'' with the actual MPG column name

# Create a boxplot to show MPG distribution by car type
plt.figure(figsize=(10, 6))
sns.boxplot(x="type", y="mpg", data=df, palette="Set2")

plt.title("MPG Distribution by Car Type")
plt.xlabel("Car Type")
plt.ylabel("Miles per Gallon (MPG)")
plt.xticks(rotation=45)  # Rotate x-labels if needed
plt.tight_layout()
plt.show()
```
